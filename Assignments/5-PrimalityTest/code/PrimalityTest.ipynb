## Primality Test - Implementazione e Analisi di n = 65
In questo laboratorio e' richiesto di verificare il comportamento dell'algoritmo $\text{MCPrimalityTest(n) su }, n=65$, verficare che $H(65)$ sia un sottogruppo di $\mathbb{Z}^*_{65}$ e di produrre un _MR-Testimone_ banale e non banale del fatto che $65$ sia un numero composto.

Inizio definendo il comportamento dell'algoritmo e richiamando alcuni concetti teorici alla base di questo. \
Il protocollo si basa su due concetti teorici chiave:
1. **Piccolo Teorema di Fermat**, il quale afferma che, considerati $p$ _primo_ e $a \in \mathbb{Z}$ non divisibile per $p$, $a^{p-1} \equiv 1 ~ (mod ~ p)$
2. **Radici quadrate dell'unita'**, ossia che, considerato $p$ _primo_, $x_+=1$ e $x_-=-1$ sono le uniche soluzioni dell'equazione $x^2 \equiv 1 ~ (mod ~ p)$

Inoltre si ha che, considerato un qualsiasi $n \geq 3$, input dell'algoritmo, puo' essere riscritto con $q$ dispari, $s \geq 1$ come $n = q \cdot 2^s+1$ e conseguentemente possiamo scomporre $a^{n-1} - 1$ in una sequenza di moltiplicazioni $(q^{q2^{s-1}}+1)(q^{q2^{s-2}}+1) ... (a^q + 1)(a^q-1) \equiv 0 ~ (mod ~ n)$. Avremo che, se $n$ e' _primo_, uno dei fattori di questa scomposizione sara' necessariamente nullo, ossia che il fattore $(X \pm 1)$ sia multiplo di $n$.  Da questo otteniamo che esisteranno solo due sequenza possibili (_quadrature_), della forma $1,...,1$, oppure $...,-1,1,...,1$. 

Infine definiamo cosa intendiamo per _MR-Testimone_.
Se per qualche _base_ $a$, l'algoritmo produce l'_output_ ```n composto```, $a$ sara' un _MR-Testimone_ per $n$ e $n$ sara' **certamente composto**. Questo accade se incontriamo una sequenza di _quadrature_ diversa da quelle precedentemente dichiarate. Se per $n$ _composto_, una _base_ $a$ produce l'_output_ ```n probabilmente primo``` , tale $a$ sara' un _MR-Bugiardo_ per $n$.\
Vengono definiti _banali_ nel caso in cui $a$ e $n$ non siano coprimi, se $MCD(a, n) > 1$. Le basi estreme, $a=1$, $a=n-1$ saranno sempre _MR-Bugiardi_ per $n$ _composto_.


``` code
MCPrimalityTest(n)
    1. s = 0
    2. q = n - 1
    3. while (q pari)
        s ← s + 1
        q ← q/2
    4. campiona a uniformemente a caso in {2,3,...,n - 2}
    5. calcola x ≡ a^q (mod n) 
    6. if x ≡ ±1
        return n probabilmente primo
    7. while s - 1 ≥ 0
        x ≡ x^2 (mod n)
        if x ≡ -1 (mod n)
            return n probabilmente primo
        s ← s - 1
    8. return n composto
```

Considerato $n = 65$, troviamo la sua scomposizione in forma $q \cdot 2^s + 1$ con $q = 1, s = 6$. Avremo che $|\mathbb{Z}^*_{65}| = 48$.  \
Verifichiamo che $H_{65} = \{1,8,14,18,47,51,57,64\}$ e' un sottogruppo proprio di $\mathbb{Z}^*_{65}$ costituito dalla coppia di _MR-Banali_ $\{1, 64\}$, dalle due coppue du _MR-Bugiardi_ $\{8, 57\}$ e $\{18,47\}$ e dalla coppia di _MR-Testimoni_ $\{14, 51\}$.

Verifichiamo, per la base $a = 2$ il funzionamento dell'algoritmo:
