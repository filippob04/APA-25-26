## Primality Test - Implementazione e Analisi di n = 65
In questo laboratorio e' richiesto di verificare il comportamento dell'algoritmo $\text{MCPrimalityTest(n) su }, n=65$, verficare che $H(65)$ sia un sottogruppo di $\mathbb{Z}^*_{65}$ e di produrre un _MR-Testimone_ banale e non banale del fatto che $65$ sia un numero _composto_.

Inizio definendo il comportamento dell'algoritmo e richiamando alcuni concetti teorici alla base di questo.
Il protocollo si basa su due concetti chiave:

1. **Piccolo Teorema di Fermat**, il quale afferma che, considerati $p$ _primo_ e $a \in \mathbb{Z}$ non divisibile per $p$, $a^{p-1} \equiv 1 ~ (mod ~ p)$

2. **Radici quadrate dell'unita'**, ossia che, considerato $p$ _primo_, $x_+=1$ e $x_-=-1$ sono le uniche soluzioni dell'equazione $x^2 \equiv 1 ~ (mod ~ p)$

Inoltre si ha che, considerato un qualsiasi $n \geq 3$, input dell'algoritmo, puo' essere riscritto con $q$ dispari, $s \geq 1$ come $n = q \cdot 2^s+1$ e conseguentemente possiamo scomporre $a^{n-1} - 1$ in una sequenza di moltiplicazioni $(a^{q2^{s-1}}+1)(a^{q2^{s-2}}+1) ... (a^q + 1)(a^q-1) \equiv 0 ~ (mod ~ n)$. Avremo che, se $n$ e' _primo_ uno dei fattori di questa scomposizione sara' necessariamente nullo, (il fattore $(X \pm 1)$ e' un multiplo di $n$).  Da questo otteniamo che esisteranno solo due sequenza possibili (_quadrature_), della forma $1,...,1$, oppure $...,-1,1,...,1$. 

Infine definiamo cosa intendiamo per _MR-Testimone_.
Se per qualche _base_ $a$, l'algoritmo produce l'_output_ ```n composto```, $a$ sara' un _MR-Testimone_ per $n$ e $n$ sara' **certamente composto**. Questo accade se incontriamo una sequenza di _quadrature_ diversa da quelle precedentemente dichiarate. Se per $n$ _composto_, una _base_ $a$ produce l'_output_ ```n probabilmente primo``` , tale $a$ sara' un _MR-Bugiardo_ per $n$.\
Vengono definiti _banali_ nel caso in cui $a$ e $n$ non siano coprimi, se $MCD(a, n) > 1$. Le basi estreme, $a=1$, $a=n-1$ saranno sempre _MR-Bugiardi_ per $n$ _composto_.

``` code
MCPrimalityTest(n)
    1. s = 0
    2. q = n - 1
    3. while (q pari)
        s ← s + 1
        q ← q/2
    4. campiona a uniformemente a caso in {2,3,...,n - 2}
    5. calcola x ≡ a^q (mod n) 
    6. if x ≡ ±1
        return n probabilmente primo
    7. while s - 1 ≥ 0
        x ≡ x^2 (mod n)
        if x ≡ -1 (mod n)
            return n probabilmente primo
        s ← s - 1
    8. return n composto
```

Mostro adesso un'implementazione dell'Algoritmo $\text{MCPrimalityTest(65)}$ per $a=2,...,63$. 

````python
# Imports
import IPython
import warnings

import random
````

````python
# MCPrimalityTest(n)
# Input: n >=3 dispari
# Output: "n composto" o "probabilmente primo"

def MCPrimalityTest(n, a):
    if n < 3 or n % 2 == 0:
        return "NON VALID INPUT"
    s = 0
    q = n - 1
    while q % 2 == 0:
        s +=1
        q = q//2
    x = pow(a, q, n)
    if x == 1 or x == n - 1:
        return "probably prime"
    while s-1 > 0:
        x = pow(x, 2, n)
        if x == n - 1:
            return "probably prime"
        s -=1
    return "composite"




# Output
n = 65
liars = []
witnesses = []

for a in range(2, n - 1):
    res = MCPrimalityTest(n, a)
    
    if res == "probably prime":
        liars.append(a)
    else:
        witnesses.append(a)

print(f"MR-Liars: {liars}")
print(f"MR-Witnesses: {witnesses}")
````

L'insieme _MR-Liars_, $H_{65}$ e' **sottogruppo proprio** di $\mathbb{Z}^*_{65}$ (ossia $|H_{65}| < |\mathbb{Z}^*_{65}|$), infatti esiste una base $\bar{a}$ per cui $\bar{a}^{64} \not\equiv 1 ~ (mod ~ 65)$ (un _MR-Testimone_). e' anche verificato il **Teorema di Lagrange** ($|\mathbb{Z}^*_{65}| = (5-1)(13-1) = 48 \rightarrow |H_{65}| \in \{d \in \mathbb{N} ~|~ 48 ~ (mod ~ d) = 0\}$), il quale afferma che al massimo la meta' delle basi apparterranno a $H_{65}$ e dunque almeno meta' delle basi saranno _MR-Testimoni_.

La complessita' dell'algoritmo, se $\text{log~n}$ sono i bit necessari per codificare un numero $n$, il costo di $k$ ripetizioni dell'algortimo e' pari a ${O(k(log~n)^3)}$.
