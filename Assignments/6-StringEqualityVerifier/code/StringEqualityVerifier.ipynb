## String Equality Verifier - Implementazione e Verifica di un esempio pratico
In questo laboratorio, considerato un file $a$ di $30$ bit e un **numero primo** casuale $p, 2 \leq p \leq 900$, e' richiesto di determinare tutti i possibili file $b per i quali, fissato $p$, l'output di $\text{MCStringEqualityVerifier(a,b)}$ e' ```input probabilmente uguali```.

Classicamente se volessimo verificare se due file, entrambi di lunghezza $\ell$, sono identici dovremmo spedire e confrontare tutti gli $\ell$ bit. Il protocollo presentato invece funziona valutandone solo $2ln\ell$. Per far cio' e' necessario introdurre il concetto di **Impronta digitale** di un file.

Considerato un **numero primo** $p, 2 \leq p \leq \ell^2$ e $n \in \mathbb{N}$, $n < 2^\ell$, definiamo l'**Impronta digitale** di $n$:
$$f(n) = n ~ (mod~p)$$
Questa avra' una dimensione ridotta ad al piu' $2ln~\ell$ bit, derivata dal numero di bit necessari a codificare $p$ in binario. Si avra' che $p \leq \ell^2, f(n) < p \rightarrow f(n) \leq \ell^2$ ossia, il **resto** del modulo di $p$ e' strettamente minore di $p$ stesso e quindi al piu' lungo $\ell^2$. \
Calcolata l'**Impronta digitale** di ciascuno dei due file, allora bastera' confrontarle fra loro per ottenere il verdetto.

Presento ora l'algoritmo $\text{MCStringEqualityVerifier(a, b)}$ che prende in _Input_ $a, b$ **stringhe binarie** di lunghezza $\ell$ e restituisce in _output_ ```input probabilmente uguali``` se le **impronte digitali** combaciano, altrimenti ```input diversi```.

```code
1. campiona p, primo, uniformemente a caso in {2, ..., l^2}
2. f(a) ← a (mod p)
3. f(b) ← b (mod p)
4. if |f(a) - f(b)| ≡ 0 (mod p)
    return input probabilmente uguali
5. else
    return input diversi
```

Il protocollo si compone delle seguenti fasi:

1) **Alice** seleziona un **numero primo** $p$ a _caso_ fra $2$ e $\ell^2$ \
2) **Alice** _calcola_ l'**impronta** $f(a) = a ~ (mod ~ p)$ del suo file **A** (lungo $\ell$ bit) e _spedisce_ sia il **numero primo** che l'**impronta** digitale a Bob. \
3) **Bob** _riceve_ $p$ e $f(a)$ e _calcola_ l'**impronta digitale** del suo file, $f(b) = b ~ (mod ~ p)$ \
4) **Bob** _confronta_ le due **impronte digitali** (i due resti), e _verifica_ se sono uguali.

I due file a questo punto vengono considerati **probabilmente identici**.
La parola **probabilmente**, come sempre sta ad indicare la possibilita' d'errore dell'algoritmo, infatti se $p$, scelto casualmente, e' un **divisore esatto** di $|f(a) - f(b)|$, avremmo comunque che $f(a) - f(b)~(mod ~ p) = 0$. La probabilita' d'_errore_ e' calcolabile.

Infatti, il numero di **fattori primi** di un numero $n < 2^\ell, n \in \mathbb{N}$ e' al piu' $\ell$ e, poiche' $p \leq \ell^2$, avremo un totale di numeri primi _possibili_ pari a $\pi(p) = \frac{\ell^2}{2ln~\ell}$ quindi:
$$\mathbb{P}(\ell) \approx \frac{\ell}{\pi(p)} \approx \frac{2ln~\ell}{\ell}$$

---

Consideriamo ora un file $a$ di $30$ bit, e $p$ primo **fissato**, $2 \leq p \leq 900$. \
Per determinare tutti i file per i quali l'output e' ```probabilmente uguali``` devo calcolare il numero di casi in cui il verdetto dell'algoritmo e' positivo.

Ricordando che un falso positivo avviene se $f(a) \equiv f(b) ~(mod ~p)$, cioe' quando $p$ e' un **fattore primo** della differenza, quindi $a - b = k \cdot p, k \in \mathbb{Z}$. \
Per $a, b \leq 2^{30}$ avremo:
$$B_\text{probabilmente uguali} = \{b \in \mathbb{N} ~|~ b = a - k \cdot p, 0 \leq b \leq 2^{30}\}$$

Quindi su $b = 2^{30} \approx 10^9$, tutti i file possibili da $30$ _bit_, scelto a _caso_ un $p = 101$, il numero di file uguali ad $a$ e' pari a $\frac{10^9}{101} \approx 10^7$.

Considerato il totale di numeri primi a disposizione $\pi(900) = \frac{900}{ln(900)} \approx 132$ La probabilita' che tale valore sia stato scelto da un altro mio collega e' $\mathbb{P}(\text{same}) = \frac{1}{132} \approx 0.76\%$.
